```python
def for_pie(x):
    j= 1
    for i in x:
        if i == True:
            j+=1
    return j, len(x)-j
```

```python
team_names = pd.unique(df['winner'])
```

```python
scores = []
for i in team_names:
    x = df['winner'] == i
    val1, val2 = for_pie(x)
    scores.append(val1)
```

```python
for i in range(len(team_names)):
    if team_names[i] is np.nan:
        team_names[i] = 'Other'
```

```python
plt.figure(figsize=(20,10))
plt.title('Most Of the Time Winners')
plt.xlabel('Teams')
plt.ylabel('COUNT')
plt.bar(team_names, scores)
plt.xticks(rotation = 45)
plt.show()
```
